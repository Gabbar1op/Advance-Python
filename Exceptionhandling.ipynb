```python
#1. Student Grade Management System with Exception Handling
students = {}

def add_student():
    try:
        student_id = input("Enter Student ID: ")
        if student_id.strip() == "":
            raise ValueError("Student ID cannot be empty!")

        grade = input("Enter Grade: ")
        if grade.strip() == "":
            raise ValueError("Grade cannot be empty!")

        grade = float(grade)   # May cause type mismatch

        students[student_id] = grade
        print("Student added successfully!")

    except ValueError as e:
        print("Error:", e)

def update_student():
    try:
        student_id = input("Enter Student ID to update: ")

        if student_id not in students:
            raise KeyError("Invalid Student ID!")

        grade = input("Enter new Grade: ")
        if grade.strip() == "":
            raise ValueError("Grade cannot be empty!")

        grade = float(grade)

        students[student_id] = grade
        print("Grade updated successfully!")

    except KeyError as e:
        print("Error:", e)
    except ValueError as e:
        print("Error:", e)

def delete_student():
    try:
        student_id = input("Enter Student ID to delete: ")

        if student_id not in students:
            raise KeyError("Invalid Student ID!")

        del students[student_id]
        print("Student record deleted!")

    except KeyError as e:
        print("Error:", e)

def display_students():
    if not students:
        print("No student records found.")
    else:
        print("\nStudent Records:")
        for sid, grade in students.items():
            print("ID:", sid, "Grade:", grade)

while True:
    print("\n--- Student Grade Management System ---")
    print("1. Add Student")
    print("2. Update Grade")
    print("3. Delete Student")
    print("4. Display Students")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        update_student()
    elif choice == "3":
        delete_student()
    elif choice == "4":
        display_students()
    elif choice == "5":
        print("Exiting program...")
        break
    else:
        print("Invalid choice! Please select 1-5.")
```

```python
#2. Contact Book with Exception Handling
contacts = {}

def add_contact():
    try:
        name = input("Enter Name: ")
        if name.strip() == "":
            raise ValueError("Name cannot be empty!")

        if name in contacts:
            raise KeyError("Contact already exists!")

        phone = input("Enter Phone Number: ")
        if phone.strip() == "":
            raise ValueError("Phone number cannot be empty!")

        if not phone.isdigit() or len(phone) != 10:
            raise ValueError("Phone number must be 10 digits!")

        contacts[name] = phone
        print("Contact added successfully!")

    except (ValueError, KeyError) as e:
        print("Error:", e)

def edit_contact():
    try:
        name = input("Enter Name to Edit: ")

        if name not in contacts:
            raise KeyError("Contact not found!")

        phone = input("Enter New Phone Number: ")

        if not phone.isdigit() or len(phone) != 10:
            raise ValueError("Invalid phone number format!")

        contacts[name] = phone
        print("Contact updated successfully!")

    except (ValueError, KeyError) as e:
        print("Error:", e)

def search_contact():
    try:
        name = input("Enter Name to Search: ")

        if name.strip() == "":
            raise ValueError("Search field cannot be empty!")

        if name not in contacts:
            raise KeyError("Contact not found!")

        print("Name:", name)
        print("Phone:", contacts[name])

    except (ValueError, KeyError) as e:
        print("Error:", e)

while True:
    print("\n--- Contact Book ---")
    print("1. Add Contact")
    print("2. Edit Contact")
    print("3. Search Contact")
    print("4. Display All")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_contact()
    elif choice == "2":
        edit_contact()
    elif choice == "3":
        search_contact()
    elif choice == "4":
        print("\nContacts:")
        for name, phone in contacts.items():
            print(name, "-", phone)
    elif choice == "5":
        print("Exiting Contact Book...")
        break
    else:
        print("Invalid choice!")
```

```python
#3. Banking System with Exception Handling
accounts = {
    "1001": 5000,
    "1002": 3000,
    "1003": 7000
}

def deposit():
    try:
        acc = input("Enter Account Number: ")

        if acc not in accounts:
            raise KeyError("Incorrect Account Number!")

        amount = float(input("Enter Amount to Deposit: "))
        accounts[acc] += amount

        print("Deposit Successful! New Balance:", accounts[acc])

    except KeyError as e:
        print("Error:", e)
    except ValueError:
        print("Invalid amount entered!")

def withdraw():
    try:
        acc = input("Enter Account Number: ")

        if acc not in accounts:
            raise KeyError("Incorrect Account Number!")

        amount = float(input("Enter Amount to Withdraw: "))

        if amount > accounts[acc]:
            raise Exception("Overdraft Error! Insufficient Balance.")

        accounts[acc] -= amount
        print("Withdrawal Successful! Balance:", accounts[acc])

    except KeyError as e:
        print("Error:", e)
    except ValueError:
        print("Invalid amount entered!")
    except Exception as e:
        print("Error:", e)

def transfer():
    try:
        sender = input("Enter Sender Account: ")
        receiver = input("Enter Receiver Account: ")

        if sender not in accounts or receiver not in accounts:
            raise KeyError("Incorrect Account Number!")

        amount = float(input("Enter Amount to Transfer: "))

        if amount > accounts[sender]:
            raise Exception("Overdraft Error! Insufficient Balance.")

        accounts[sender] -= amount
        accounts[receiver] += amount

        print("Transaction Successful!")

    except KeyError as e:
        print("Error:", e)
    except ValueError:
        print("Invalid amount!")
    except Exception as e:
        print("Error:", e)

def display_accounts():
    print("\nAccount Balances:")
    for acc, balance in accounts.items():
        print("Account:", acc, "Balance:", balance)

while True:
    print("\n--- Banking System ---")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Transfer Money")
    print("4. Display Accounts")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        deposit()
    elif choice == "2":
        withdraw()
    elif choice == "3":
        transfer()
    elif choice == "4":
        display_accounts()
    elif choice == "5":
        print("Exiting Banking System...")
        break
    else:
        print("Invalid choice!")
```

```python
#4. E-Commerce Order Management System with Exception Handling
products = {
    "Laptop": {"price": 50000, "stock": 5},
    "Phone": {"price": 20000, "stock": 10},
    "Headphones": {"price": 2000, "stock": 8}
}

valid_coupon = "SAVE10"
orders = []

def place_order():
    try:
        product = input("Enter Product Name: ")

        if product not in products:
            raise KeyError("Product not available!")

        quantity = int(input("Enter Quantity: "))

        if quantity > products[product]["stock"]:
            raise Exception("Out-of-stock error!")

        coupon = input("Enter Coupon Code (or press Enter): ")
        price = products[product]["price"] * quantity

        if coupon:
            if coupon != valid_coupon:
                raise ValueError("Invalid Coupon Code!")
            price = price * 0.9   # 10% discount

        payment = input("Enter Payment Method (UPI/Card/Cash): ")

        if payment not in ["UPI", "Card", "Cash"]:
            raise ValueError("Invalid Payment Method!")

        products[product]["stock"] -= quantity
        orders.append({"product": product, "quantity": quantity, "price": price})

        print("Order placed successfully! Total Price:", price)

    except KeyError as e:
        print("Error:", e)
    except ValueError as e:
        print("Error:", e)
    except Exception as e:
        print("Error:", e)

def return_order():
    try:
        product = input("Enter Product to Return: ")

        for order in orders:
            if order["product"] == product:
                products[product]["stock"] += order["quantity"]
                print("Return successful!")
                return

        raise KeyError("Order not found!")

    except KeyError as e:
        print("Error:", e)

def refund_order():
    try:
        product = input("Enter Product for Refund: ")

        for order in orders:
            if order["product"] == product:
                print("Refund Amount:", order["price"])
                return

        raise KeyError("Invalid Order!")

    except KeyError as e:
        print("Error:", e)

while True:
    print("\n--- E-Commerce Order System ---")
    print("1. Place Order")
    print("2. Return Order")
    print("3. Refund")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        place_order()
    elif choice == "2":
        return_order()
    elif choice == "3":
        refund_order()
    elif choice == "4":
        print("Exiting system...")
        break
    else:
        print("Invalid choice!")
```

```python
#5. Custom Exception Classes
class OutOfStockError(Exception):
    pass

class InvalidProductIDError(Exception):
    pass

class InvalidQuantityError(Exception):
    pass

# Inventory Data
inventory = {
    "P101": {"name": "Laptop", "stock": 5},
    "P102": {"name": "Phone", "stock": 10},
    "P103": {"name": "Headphones", "stock": 7}
}

def purchase_product():
    try:
        pid = input("Enter Product ID: ")

        if pid not in inventory:
            raise InvalidProductIDError("Product ID does not exist!")

        quantity = int(input("Enter Quantity: "))

        if quantity <= 0:
            raise InvalidQuantityError("Quantity must be greater than 0!")

        if quantity > inventory[pid]["stock"]:
            raise OutOfStockError("Requested quantity not available!")

        inventory[pid]["stock"] -= quantity

        print("Purchase successful!")
        print("Remaining stock:", inventory[pid]["stock"])

    except InvalidProductIDError as e:
        print("Error:", e)

    except OutOfStockError as e:
        print("Error:", e)

    except InvalidQuantityError as e:
        print("Error:", e)

    except ValueError:
        print("Error: Quantity must be a number!")

def display_inventory():
    print("\nInventory Details:")
    for pid, details in inventory.items():
        print(pid, "-", details["name"], "| Stock:", details["stock"])

while True:
    print("\n--- Inventory Management System ---")
    print("1. Purchase Product")
    print("2. Display Inventory")
    print("3. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        purchase_product()
    elif choice == "2":
        display_inventory()
    elif choice == "3":
        print("Exiting program...")
        break
    else:
        print("Invalid choice!")
```

```python
#6. Flight Booking System with Exception Handling
flights = {
    "F101": {"destination": "Delhi", "seats": 5, "price": 5000},
    "F102": {"destination": "Mumbai", "seats": 3, "price": 4500},
    "F103": {"destination": "Bangalore", "seats": 4, "price": 6000}
}

bookings = []

def search_flights():
    print("\nAvailable Flights:")
    for fid, details in flights.items():
        print(fid, "| Destination:", details["destination"],
              "| Seats:", details["seats"], "| Price:", details["price"])

def book_ticket():
    try:
        fid = input("Enter Flight ID: ")

        if fid not in flights:
            raise KeyError("Invalid Flight ID!")

        name = input("Enter Passenger Name: ")
        if name.strip() == "":
            raise ValueError("Invalid passenger details!")

        if flights[fid]["seats"] <= 0:
            raise Exception("Seat not available!")

        payment = input("Enter Payment Method (Card/UPI): ")
        if payment not in ["Card", "UPI"]:
            raise Exception("Payment failure! Invalid method.")

        flights[fid]["seats"] -= 1
        bookings.append({"flight": fid, "name": name})

        print("Ticket booked successfully!")

    except KeyError as e:
        print("Error:", e)
    except ValueError as e:
        print("Error:", e)
    except Exception as e:
        print("Error:", e)

def cancel_ticket():
    try:
        name = input("Enter Passenger Name to Cancel Ticket: ")

        for booking in bookings:
            if booking["name"] == name:
                fid = booking["flight"]
                flights[fid]["seats"] += 1
                bookings.remove(booking)
                print("Ticket cancelled successfully!")
                return

        raise KeyError("Booking not found!")

    except KeyError as e:
        print("Error:", e)

while True:
    print("\n--- Flight Booking System ---")
    print("1. Search Flights")
    print("2. Book Ticket")
    print("3. Cancel Ticket")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        search_flights()
    elif choice == "2":
        book_ticket()
    elif choice == "3":
        cancel_ticket()
    elif choice == "4":
        print("Exiting system...")
        break
    else:
        print("Invalid choice!")
```

```python
#7. Exception division by 3 error
class NotDivisibleByThreeError(Exception):
    pass

try:
    num = int(input("Enter a number: "))

    if num % 3 != 0:
        raise NotDivisibleByThreeError("Number is not divisible by 3")

except NotDivisibleByThreeError as e:
    print("Custom Exception:", e)

except ValueError:
    print("Invalid input! Please enter an integer.")

else:
    result = num / 3
    print("Division result:", result)

finally:
    print("This block always executes.")
```
